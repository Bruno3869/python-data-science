```python
import csv
f = open("nfl_suspensions_data.csv")
nfl_suspensions = list(csv.reader(f))
nfl_suspensions = nfl_suspensions[1:]

print(nfl_suspensions[0:5])
```

```python
years = {}

for suspension in nfl_suspensions:
    row_year = suspension[5]
    if row_year in years:
        years[row_year] += 1
    else:
        years[row_year] = 1

print(years)
```

## Valeurs uniques

```python
#Bruno
unique_teams = []
unique_games = []

for team in nfl_suspensions:
    row_team = team[1]
    unique_teams.append(row_team)
    
for game in nfl_suspensions:
    row_game = game[2]
    unique_games.append(row_game)
    
print(set(unique_teams))
print(set(unique_games))
```

```python
#Rod
teams = [row[1] for row in nfl_suspensions]
unique_teams = set(teams)
print(unique_teams)
```

```python
#Rod
games = [row[2] for row in nfl_suspensions]
unique_games = set(games)
print(unique_games)
```

## Am√©lioration de la classe Suspension

```python
class Suspension():  
    def __init__(self, row):
        self.name = row[0]
        self.team = row[1]
        self.games = row[2]
        try:
            self.year = int(row[5])
        except Exception:
            self.year = 0
            
    def get_year(self):
        return(self.year)
```

```python
missing_year = Suspension(nfl_suspensions[6])
```

```python
get_missing_year = missing_year.get_year()
```
