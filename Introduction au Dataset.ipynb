## 2. Convertir la data en liste de listes

* Créer une fonction read_csv() qui à partir du fichier "US_births_2000-2014.csv" retourne une liste de listes dans laquelle toutes les valeurs sont des entiers
* Assigner le résultat à us_births_list
* Afficher les 10 premiers éléments

```python
def read_csv(file):
    f = open(file, "r")
    string_list = f.read()
    string_list = string_list.split("\n")
    string_list = string_list[1:len(string_list)-1]
    
    final_list = []
    
    for row in string_list:
        int_fields = []
        string_fields = row.split(",")
        for row in string_fields:
            int_fields.append(int(row))
        final_list.append(int_fields)
    return(final_list)
    
us_births_list = read_csv("US_births_2000-2014.csv")

print(us_births_list[0:10])
```

## 3. Calculer le nombre de naissances par mois

* Créer une fonction month_births() qui à partir de la liste de liste us_births_list retourne un dictionnaire contenant pour chaque mois le nombre de naissances
* Assigner le résultat à us_month_births
* Afficher le résultat

```python
def month_births(list):
    births_per_month = {}
    
    for li in list:
        month = li[1]
        births = li[4]
        
        if month in births_per_month:
            births_per_month[month] += births
        else:
            births_per_month[month] = births
        
    return(births_per_month)

us_month_births = month_births(us_births_list)
print(us_month_births)
```

## 4. Calculer le nombre de naissances par jours de la semaine

* Créer une fonction day_births() qui à partir de la liste de liste us_births_list retourne le nombre de naissance pour chaque jour de la semaine
* Assigner le résultat à us_days_births
* Afficher le résultat

```python
def day_births(list):
    births_per_day = {}
    
    for li in list:
        day = li[3]
        births = li[4]
        
        if day in births_per_day:
            births_per_day[day] += births
        else:
            births_per_day[day] = births
    
    return(births_per_day)

us_days_births = day_births(us_births_list)
print(us_days_births)
```

## 5. Créer une fonction plus générale

* Créer une fonction plus générale calc_counts() qui à partir de la liste de listes us_births_list et du numéro de colonne (0 -> année, 1 -> mois et 3 -> jour de la semaine) retourne le nombre de naissances pour la colonne choisie
* Tester la fonction avec les années et assigner le résultat à us_year_births

```python
def calc_counts(list, column):
    births_per_col = {}
    
    for li in list:
        key = li[column]
        births = li[4] 
        
        if key in births_per_col:
            births_per_col[key] += births
        else:
            births_per_col[key] = births
    
    return(births_per_col)

us_year_births = calc_counts(us_births_list, 0) #0 -> année, 1 -> mois et 3 -> jour de la semaine
print(us_year_births)
```
