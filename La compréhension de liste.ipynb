## Fonction enumerate()

```python
students = ["Anna", "Martin", "Bob", "Eric", "Elise"]
ages = [16, 15, 18, 20, 20]
```

```python
for student in students:
    print("Etudiant")
    print(student)
    print("Age")
```

```python
for i, student in enumerate(students):
    print("Index Etudiant")
    print(i)
    print("Etudiant")
    print(student)
```

```python
for i, student in enumerate(students):
    print("Index Etudiant")
    print(i)
    print("Etudiant")
    print(student)
    print("Age")
    print(ages[i])
```

```python
cars = [["Noir", "Peugeot", "308"], ["Blanche", "Audi", "A4"]]
price = [23000, 49000]
```

```python
for i, car in enumerate(cars):
    car.append(price[i])
```
