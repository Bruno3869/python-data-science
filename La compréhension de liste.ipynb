## Fonction enumerate()

```python
students = ["Anna", "Martin", "Bob", "Eric", "Elise"]
ages = [16, 15, 18, 20, 20]
```

```python
for student in students:
    print("Etudiant")
    print(student)
    print("Age")
```

```python
for i, student in enumerate(students):
    print("Index Etudiant")
    print(i)
    print("Etudiant")
    print(student)
```

```python
for i, student in enumerate(students):
    print("Index Etudiant")
    print(i)
    print("Etudiant")
    print(student)
    print("Age")
    print(ages[i])
```

```python
cars = [["Noir", "Peugeot", "308"], ["Blanche", "Audi", "A4"]]
price = [23000, 49000]
```

```python
for i, car in enumerate(cars):
    car.append(price[i])
```

## Compréhension de liste

```python
animals = ["Chien", "Tigre", "Lion", "Vache", "Panda"]

animal_lengths = []
for animal in animals:
    animal_lengths.append(len(animal))
    
print(animal_lengths)
```

```python
animal_lengths_2 = [len(animal) for animal in animals]
print(animal_lengths_2)
```

```python
prices = [10, 45, 156, 7800]
prices_doubled = [price * 2 for price in prices]
print(prices_doubled)
```

### Training

```python
import csv
f = open("legislators.csv")
legislators = list(csv.reader(f))

birth_year = []

for row in legislators:
    birthday = row[2]
    birth_year = birthday.split("-")[0]
    try:
        birth_year = int(birth_year)
    except Exception:
        birth_year = 0
    row.append(birth_year) #ajout d'une nouvelle colonne

legislators[0][7] = "birth_year" #ajout d'un nouvelle entête de colonne index0 colonne7

print(legislators[0:10])
```

```python
name_counts = {}

for row in legislators:
    gender = row[3]
    year = row[7]
    
    if gender == "F" and year > 1950:
        name = row[1]
        if name in name_counts:
            name_counts[name] += 1
        else:
            name_counts[name] = 1

print(name_counts)
```

```python
values = [5, 70, 110]
max_value = 0
for i in values:
    if i > max_value:
        max_value = i

print(max_value)
```

```python
values = [-5, -70, -110]
max_value = 0
for i in values:
    if i > max_value:
        max_value = i

print(max_value)
```

```python
values = [-5, -70, -110]
max_value = None
for i in values:
    if max_value is None or i > max_value:
        max_value = i

print(max_value)
```

### Training

```python
male_name_counts = {}

for row in legislators:
    gender = row[3]
    year = row[7]
    
    if gender == "M" and year > 1940:
        name = row[1]
        if name in male_name_counts:
            male_name_counts[name] += 1
        else:
            male_name_counts[name] = 1

print(male_name_counts)
```

```python
top_male_count = None

for name, count in male_name_counts.items():
    if top_male_count is None or count > top_male_count:
        top_male_count = count

print(top_male_count)
```

```python
top_male_names = []

for name, count in male_name_counts.items():
    if count == top_male_count:
        top_male_names.append(name)
        
print(top_male_names)
```
