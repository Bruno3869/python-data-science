## Qu'est-ce qu'un dictionnaire

```python
students = ["Thomas", "Matt", "Sarah", "Nina"]
scores = [16, 15, 17, 14]

index = [0, 1, 2, 3]
name = "Sarah"
score = 0

for i in index:
    if students[i] == name:
        score = scores[i]
        
print(score)
```

## La condition if / else

```python
temperature = 50

if temperature > 40:
    print("Il fait chaud !")
if temperature <= 40:
    print("Il fait froid !")
```

```python
temperature = 10

if temperature > 40:
    print("Il fait chaud !")
else:
    print("Il fait froid !")
```

```python
temperatures = [50, 20, 10, 35]

high_temperatures = []
low_temperatures = []

for temperature in temperatures:
    if temperature > 40:
        high_temperatures.append(temperature)
    else:
        low_temperatures.append(temperature)
        
print(high_temperatures)
print(low_temperatures)
```

### Trainning

```python
planet_names = ["Mercure", "Vénus", "Terre", "Mars", "Jupiter", "Saturne", "Neptune", "Uranus"]

long_names = []
short_names = []

for planet_name in planet_names:
    if len(planet_name) > 5:
        long_names.append(planet_name)
    else:
        short_names.append(planet_name)
        
print(long_names)
print(short_names)
```

## Compter les éléments d'une liste et présenter les résultats dans un dictionnaire

```python
fruits = ["Citron", "Banane", "Pomme", "Pomme", "Pomme", "Banane", "Banane", "Pomme"]

fruits_count = {}

for item in fruits:
    if item in fruits_count:
        fruits_count[item] = fruits_count[item] + 1
    else:
        fruits_count[item] = 1
        
print(fruits_count)
```
