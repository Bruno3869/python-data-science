# Les Dictionnaires

## Qu'est-ce qu'un dictionnaire

```python
students = ["Thomas", "Matt", "Sarah", "Nina"]
scores = [16, 15, 17, 14]

index = [0, 1, 2, 3]
name = "Sarah"
score = 0

for i in index:
    if students[i] == name:
        score = scores[i]
        
print(score)
```

```python
scores = {} #Dictionnaire vide
```

```python
scores ["Nina"] = 14
```

```python
scores ["Thomas"] = 16
scores ["Matt"] = 15
scores ["Sarah"] = 17
```
