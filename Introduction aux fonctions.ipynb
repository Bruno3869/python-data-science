### Training

```python
def clean_text(string_value):
    cleaned_value = string_value.replace(".", "")
    cleaned_value = string_value.replace(",", "")
    cleaned_value = string_value.replace("'", "")
    cleaned_value = string_value.replace("\n", "")
    return(cleaned_value)

f = open("texte.txt", "r", encoding="utf-8")
text_string = f.read()

text_string = clean_text(text_string)

print(text_string)
```
