# Challenge Département Francais

#### Ouvrir un nouveau fichier et créer une liste d'entier comprenant le nombre d'habitants de tous les départements français.

```python
f = open("departements_fr.csv", "r", encoding="UTF-8")
data = f.read()
rows = data.split("\n")

list_data = []

for row in rows:
    split_list = row.split(",")
    list_data.append(split_list)

population = list_data[0:len(list_data)-1]
  
department_populations = []

for row in population:
    department_population = int(row[1])
    department_populations.append(department_population)
    
print(department_populations)

print(type(department_populations[0]))
```
