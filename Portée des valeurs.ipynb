## Les portées des variables

```python
def add(a, b):
    total = a + b
    return total

total = 15
```

```python
print(add(10, 20))
```

```python
print(total)
```

```python
#Variables Locales
def add(a, b):
    total = a + b
    return total

def subtract(a, b):
    total = a - b
    return total
```

```python
print(add(1, 5))
```

```python
print(subtract(1, 5))
```

```python
#Variables Globales
total = 10

def add_to_total(a):
    global total
    total = total + a
```

```python
print(total)
```

```python
add_to_total(20)
```

```python
print(total)
```
