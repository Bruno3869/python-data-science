```python
f = open("departements_fr.csv", "r")
data = f.read()
data_clean = data[0:len(data)-1]
rows = data_clean.split("\n")

departments = []
int_dep_populations = []

for row in rows:
    values = row.split(",")
    departments.append(values[0])
    int_dep_populations.append(int(values[1]))

print(departments)
print(int_dep_populations)
```
