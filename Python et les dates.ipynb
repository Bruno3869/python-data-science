# Python et les dates

## Le module Time

```python
# time.time() le nombre de seconde depuis 1970

import time
current_time = time.time()

print(current_time)
```

```python
# time.gmtime()

current_struct_time = time.gmtime()

print(current_struct_time)
```

```python
current_hour = current_struct_time.tm_hour #affiche l'heure exacte

print(current_hour)
```

```python
current_year = current_struct_time.tm_year #affiche l'heure exacte

print(current_year)
```

## Le module datetime

```python
# datetime.now()

import datetime

current_datetime = datetime.datetime.now()
current_year = current_datetime.year
current_month = current_datetime.month

print(current_year)
print(current_month)
print(current_datetime)
```

## Classe timedelta

```python
today = datetime.datetime.now()
```

```python
diff = datetime.timedelta(weeks = 1, days = 5)
```

```python
print(today)
```

```python
print(diff)
```

```python
result = today + diff
```

```python
print(result)
```

```python
diff = datetime.timedelta(days = 1)
```

```python
tomorrow = today + diff
yesterday = today - diff
```

```python
print(tomorrow)
```

```python
print(yesterday)
```

## Formater nos dates

```python
today
```

```python
# datetime.strftime()

string_today = today.strftime("%b %d %Y")
```

```python
print(string_today)
```

```python
# strptime()

today2 = time.strptime(string_today, "%b %d %Y") # pour reconvertir dans un format date
```
