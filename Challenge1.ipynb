```python
f = open("unisex_names.csv", "r")
names = f.read()
names_clean = names[0:len(names)-1]
names_list = names_clean.split("\n")
print (names_list[0:10])
```

### Convertir la liste en liste de liste

```python
names_data = []

for name in names_list:
    comma_list = name.split(",")
    names_data.append(comma_list)
    
print (names_data[0:10])
```

### Convertir les valeurs en numÃ©rique

```python
numerical_list = []

for line in names_data:
    name = line[0]
    count = float(line[1])
    new_list = [name, count]
    numerical_list.append(new_list)
    
print(numerical_list[0:10])
```

### Filtrer la liste

```python
final_list = []

for line in numerical_list:
    name = line[0]
    population = line[1]
    if line[1] > 1000:
        final_list.append(name)

print(final_list[0:10])
```
