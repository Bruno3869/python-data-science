# Objets et Classes

```python
class Car():
    def __init__(self):
        self.color = "Black"
        self.mark = "Audi"
        self.model = "A4"
```

```python
audi_car = Car()
```

```python
class Team():
    def __init__(self):
        self.name = "Manchester"
```

```python
man = Team()
```

```python
print(man.name)
```

## La m√©thode d'instance __init__

```python
class Team():
    def __init__(self, name):
        self.name = name
```

```python
madrid = Team("Real Madrid")
```

```python
print(madrid.name)
```
