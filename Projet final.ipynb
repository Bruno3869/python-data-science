## Nombre de décès par armes à feu aux USA par année entre 2012 et 2014

```python
years = [row[1] for row in data]
year_counts = {}
for year in years:
    if year not in year_counts: #on ajoute l'année à la liste year_counts
        year_counts[year] = 0 #les 3 années sont ajoutés avec la valeur 0
    year_counts[year] += 1
print(year_counts)
```

## Explorer les décès par armes à feu par mois et par année

```python
#Bruno
import datetime
years = []
months = []
dates = []

for row in data:
    year = int(row[1])
    month = int(row[2])
    date = datetime.datetime(year, month, day=1)
    dates.append(date)
    
print(dates[0:5])

date_count = 0
for row in dates:
    if row == row:
        date_count += 1
        
print(date_count)
```

## Explorer les décès par armes à feu par origine et par sexe

## Filtrer par homicide

```python
intents = [row[3] for row in data]
races = [row[7] for row in data]
homicide_count = {}

for i, race in enumerate(races):
    if race not in homicide_count: #on ajoute la race à la liste homicide_count
        homicide_count[race] = 0 #les races sont ajoutées avec la valeur 0
    if intents[i] == "Homicide":
        homicide_count[race] +=1

homicide_count
```

```python
homicide_ratio = {}

for k, v in homicide_count.items():
    homicide_ratio[k] = (v / mapping[k]) * 100000 #valeur de mapping qui correspond à la clé k
homicide_ratio
```
