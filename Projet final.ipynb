## Nombre de décès par armes à feu aux USA par année entre 2012 et 2014

```python
years = [row[1] for row in data]
year_counts = {}
for year in years:
    if year not in year_counts: #on ajoute l'année à la liste year_counts
        year_counts[year] = 0 #les 3 années sont ajoutés avec la valeur 0
    year_counts[year] += 1
print(year_counts)
```
