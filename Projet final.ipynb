## Nombre de décès par armes à feu aux USA par année entre 2012 et 2014

```python
years = [row[1] for row in data]
year_counts = {}
for year in years:
    if year not in year_counts: #on ajoute l'année à la liste year_counts
        year_counts[year] = 0 #les 3 années sont ajoutés avec la valeur 0
    year_counts[year] += 1
print(year_counts)
```

## Explorer les décès par armes à feu par mois et par année

```python
#Bruno
import datetime
years = []
months = []
dates = []

for row in data:
    year = int(row[1])
    month = int(row[2])
    date = datetime.datetime(year, month, day=1)
    dates.append(date)
    
print(dates[0:5])

date_count = 0
for row in dates:
    if row == row:
        date_count += 1
        
print(date_count)
```

```python
#Rod
import datetime
dates = [datetime.datetime(year = int(row[1]), month = int(row[2]), day = 1) for row in data]
dates[:5]
```

```python
#Rod
date_counts = {}
for date in dates:
    if date not in date_counts: #on ajoute la date à la liste date_counts
        date_counts[date] = 0 #les dates sont ajoutés avec la valeur 0
    date_counts[date] += 1
    
date_counts
```

## Explorer les décès par armes à feu par origine et par sexe

```python
sexes = [row[5] for row in data]
sex_counts = {}
for sex in sexes:
    if sex not in sex_counts: #on ajoute le sexe à la liste sex_counts
        sex_counts[sex] = 0 #les 2 sexes sont ajoutés avec la valeur 0
    sex_counts[sex] += 1
sex_counts
```

```python
races = [row[7] for row in data]
race_counts = {}
for race in races:
    if race not in race_counts: #on ajoute la race à la liste race_counts
        race_counts[race] = 0 #les races sont ajoutées avec la valeur 0
    race_counts[race] += 1
race_counts
```
