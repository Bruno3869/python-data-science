```python
import csv
f = open("legislators.csv")
legislators = list(csv.reader(f))

birth_year = []

for row in legislators:
    birthday = row[2]
    birth_year = birthday.split("-")[0]
    try:
        birth_year = int(birth_year)
    except Exception:
        birth_year = 0
    row.append(birth_year) #ajout d'une nouvelle colonne

legislators[0][7] = "birth_year" #ajout d'un nouvelle entête de colonne index0 colonne7

print(legislators[0:10])
```
