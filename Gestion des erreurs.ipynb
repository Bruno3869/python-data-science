# Gestion des erreurs

## Sets

```python
# set()
animals = ["Chien", "Tigre", "Chat", "Chat", "Chien", "Chien"]
unique_animals = set(animals)
print(unique_animals)
```

```python
# add()
unique_animals.add("Tortue")
print(unique_animals)
```

```python
# remove()
unique_animals.remove("Chat")
print(unique_animals)
```

```python
# list()
unique_animals = list(unique_animals)
print(unique_animals)
```
